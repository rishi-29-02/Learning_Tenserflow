## The Hello World of Deep Learning with Neural Networks

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
Xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
Ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
model.fit(Xs, Ys, epochs=500)
```

```python
print(model.predict([10.0]))
```

You might have thought 19, right? But it ended up being a little under. Why do you think that is? 

Remember that neural networks deal with probabilities, so given the data that we fed the NN with, it calculated that there is a very high probability that the relationship between X and Y is Y=2X-1, but with only 6 data points we can't know for sure. As a result, the result for 10 is very close to 19, but not necessarily 19. 

As you work with neural networks, you'll see this pattern recurring. You will almost always deal with probabilities, not certainties, and will do a little bit of coding to figure out what the result is based on the probabilities, particularly when it comes to classification.

## Improving the Fashion MNIST Accuracy using the Convnet 

```python
import tensorflow as tf
```

```python
mnist = tf.keras.datasets.fashion_mnist
```

```python
(training_images, training_labels), (test_images, test_labels) = mnist.load_data()
```

```python
print(training_images.shape)
print(test_images.shape)
```

```python
# Respaping the data
training_images = training_images.reshape(60000, 28, 28, 1)
test_images = test_images.reshape(10000, 28, 28, 1)
```

```python
# Normalize the data
training_images = training_images/255.0
test_images = test_images/255.0
```

**With Single convolution layer**

```python
cnn_model = tf.keras.Sequential([tf.keras.layers.Conv2D(64, (3, 3), activation='relu', input_shape = (28, 28, 1)),
                                 tf.keras.layers.MaxPool2D(2, 2),
                                 tf.keras.layers.Flatten(),
                                 tf.keras.layers.Dense(128, activation=tf.nn.relu),
                                 tf.keras.layers.Dense(10, activation=tf.nn.softmax)])
```

```python
cnn_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
cnn_model.fit(training_images, training_labels, epochs=5)
```

```python
cnn_model.summary()
```

```python
cnn_model.evaluate(test_images, test_labels)
```

Thus the training accuracy is 0.9057 and the test accuracy is 0.8972.


**With two convolutional layer**

```python
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if logs.get('accuracy')>0.9:
      print('Our model had reached 90% accuracy!')
      self.model.stop_training = True
```

```python
callbacks = myCallback()
```

```python
cnn_model_1 = tf.keras.Sequential([tf.keras.layers.Conv2D(64, (3, 3), activation=tf.nn.relu, input_shape=(28, 28, 1)),
                                   tf.keras.layers.MaxPool2D(2, 2),
                                   tf.keras.layers.Conv2D(64, (3, 3), activation=tf.nn.relu),
                                   tf.keras.layers.MaxPool2D(2, 2),
                                   tf.keras.layers.Flatten(),
                                   tf.keras.layers.Dense(128, activation=tf.nn.relu),
                                   tf.keras.layers.Dense(10, activation='softmax')])
```

```python
cnn_model_1.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
cnn_model_1.fit(training_images, training_labels, epochs=5, callbacks=[callbacks])
```

```python
cnn_model_1.evaluate(test_images, test_labels)
```

Thus the training accuracy is 0.9497 and the test accuracy is 0.9086.

## Exploring how convolution and Pooling works

```python
from scipy import misc
import matplotlib.pyplot as plt

i = misc.ascent()

plt.figure(figsize=(6, 6))
plt.imshow(i)
plt.gray()
plt.axis(False)
plt.show()
```

```python
print(i)
```

```python
type(i)
```

```python
i.shape
```

The image is stired in the form of a numpy array

Now lets create a convolution

```python
import numpy as np

i_transformed = np.copy(i)
size_x = i_transformed.shape[0]
size_y = i_transformed.shape[1]
```

```python
# Different filters
filter_1 = [[0, 1, 0], [1, -4, 1], [0, 1, 0]]

filter_2 = [[-1, -2, -1], [0, 0, 0], [1, 2, 1]]

filter_3 = [[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]]
weight = 1
```

The sum of all the elements in the filters should sum up to either 0 or 1.

Filter 1

```python
for x in range(size_x-1):
  for y in range(size_y-1):
    convolution = 0.0
    convolution = convolution + (i[x-1, y-1] * filter_1[0][0])
    convolution = convolution + (i[x, y-1] * filter_1[0][1])
    convolution = convolution + (i[x+1, y-1] * filter_1[0][2])
    convolution = convolution + (i[x-1, y] * filter_1[1][0])
    convolution = convolution + (i[x, y] * filter_1[1][1])
    convolution = convolution + (i[x+1, y] * filter_1[1][2])
    convolution = convolution + (i[x-1, y+1] * filter_1[2][0])
    convolution = convolution + (i[x, y+1] * filter_1[2][1])
    convolution = convolution + (i[x+1, y+1] * filter_1[2][2])
    convolution = convolution * weight
    if convolution<0:
      convolution = 0
    if convolution>255:
      convolution = 255
    i_transformed[x, y] = convolution
```

```python
plt.figure(figsize=(6, 6))
plt.imshow(i_transformed)
plt.show()
```

Filter 2

```python
for x in range(size_x-1):
  for y in range(size_y-1):
    convolution = 0.0
    convolution = convolution + (i[x-1, y-1] * filter_2[0][0])
    convolution = convolution + (i[x, y-1] * filter_2[0][1])
    convolution = convolution + (i[x+1, y-1] * filter_2[0][2])
    convolution = convolution + (i[x-1, y] * filter_2[1][0])
    convolution = convolution + (i[x, y] * filter_2[1][1])
    convolution = convolution + (i[x+1, y] * filter_2[1][2])
    convolution = convolution + (i[x-1, y+1] * filter_2[2][0])
    convolution = convolution + (i[x, y+1] * filter_2[2][1])
    convolution = convolution + (i[x+1, y+1] * filter_2[2][2])
    convolution = convolution * weight
    if convolution<0:
      convolution = 0
    if convolution>255:
      convolution = 255
    i_transformed[x, y] = convolution
```

```python
plt.figure(figsize=(6, 6))
plt.imshow(i_transformed)
plt.show()
```

Filter 3

```python
for x in range(size_x-1):
  for y in range(size_y-1):
    convolution = 0.0
    convolution = convolution + (i[x-1, y-1] * filter_3[0][0])
    convolution = convolution + (i[x, y-1] * filter_3[0][1])
    convolution = convolution + (i[x+1, y-1] * filter_3[0][2])
    convolution = convolution + (i[x-1, y] * filter_3[1][0])
    convolution = convolution + (i[x, y] * filter_3[1][1])
    convolution = convolution + (i[x+1, y] * filter_3[1][2])
    convolution = convolution + (i[x-1, y+1] * filter_3[2][0])
    convolution = convolution + (i[x, y+1] * filter_3[2][1])
    convolution = convolution + (i[x+1, y+1] * filter_3[2][2])
    convolution = convolution * weight
    if convolution<0:
      convolution = 0
    if convolution>255:
      convolution = 255
    i_transformed[x, y] = convolution
```

```python
plt.figure(figsize=(6, 6))
plt.imshow(i_transformed)
plt.show()
```

Pooling

```python
new_x = int(size_x/2)
new_y = int(size_y/2)
newImage = np.zeros((new_x, new_y))
for x in range(0, size_x, 2):
  for y in range(0, size_y, 2):
    pixels = []
    pixels.append(i_transformed[x, y])
    pixels.append(i_transformed[x, y+1])
    pixels.append(i_transformed[x+1, y])
    pixels.append(i_transformed[x+1, y+1])
    newImage[int(x/2), int(y/2)] = max(pixels)

plt.figure(figsize=(6, 6))
plt.imshow(newImage)
plt.show()
```
